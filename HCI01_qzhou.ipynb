```python
def zoom(im,x,y,bb):
    '''zoom - returns a square 2D image cut out from im centered on location (x,y) and with half-width bb'''
    return(im[y-bb:y+bb,x-bb:x+bb])

def rotate2(img, angle, c_in):
    '''rotate2 - rotate input image img by angle degrees about a point c_in
    c_in is an np.array((y,x))
    returns the rotated image with zeroes for unknown values'''
    from scipy.ndimage.interpolation import affine_transform
    # convert angle from degrees to radians
    a=angle*np.pi/180.0
    # build a rotation matrix
    transform=np.array([[np.cos(a),-np.sin(a)],[np.sin(a),np.cos(a)]])
    # work out the offset in pixels for the output image
    offset=c_in-c_in.dot(transform)
    # pass all these values into the affine transform that includes rotation and translation
    dst=affine_transform(img,transform.T,order=2,offset=offset,output_shape=(img.shape),cval=0.0)
    return(dst)
```

<b>Produce an unsharp circular mask (11 pixels in diameter) for the Beta Pic image using a convolution with a Top Hat Kernel.</b>

<div class="alert alert-block alert-info">
<b>Steps:</b> 
<ol> <li> Make circle shaped mask with 11 pixels in diameter, using the <a href="https://docs.astropy.org/en/stable/api/astropy.convolution.Tophat2DKernel.html" > <code>Tophat2DKernel</code> function</a> that is imported below</li>
    <li> Normalise the total flux in the image to 1. <i>This will be your <u>Top Hat smoothing kernel</u></i>. </li>
    <li> Convolve the Beta Pic image with this kernel to produce an image that only follows the low spatial frequencies. </li>
    <li> Subtract this smoothed image off of the original Beta Pic image. </li>
    <li> Write it out to a <code>FITS</code> file, and look at it in <code>ds9</code>.</li>

<div class="alert alert-block alert-info">
<b>For Your Information:</b> <code>usm</code> stands for the UnSharpMasking
</div>

---

```python
from astropy.convolution import convolve, Tophat2DKernel

# Q3 answer here
# creating a tophat kernel with a diameter of 11 pixels
tophat_kernel = Tophat2DKernel(11/2)
# normalising the total flux of the kernel to 1
norm = np.sum(np.array(tophat_kernel))
tophat_kernel = np.array(tophat_kernel)/norm
# getting smoothed image using convolution
smooth_pic1 = convolve(pic1,tophat_kernel)
# subtract smoothed image off of the origiinal one
usm = pic1 - smooth_pic1

# displaying the subtracted smoothed iamge
plt.imshow(zoom(usm, int(xcen), int(ycen), 20))
plt.xlabel('pixel')
plt.ylabel('pixel')
plt.show()

# writing smoothed image to file
filename = 'test_smooth.fits'
hdu = fits.PrimaryHDU(usm)
hdu.writeto(filename, overwrite=True)

# ds9(test_smooth.fits)

# END Q3 answer above

# writing subtracted smoothed image to file
filename = 'usm_tophat.fits'
hdu = fits.PrimaryHDU(usm)
hdu.writeto(filename, overwrite=True)
```

```python
from astropy.convolution import Gaussian2DKernel
from matplotlib.pyplot import figure,show
import  matplotlib

# write Q4 answer and smoothing for Kernel of 1.5 here

fig = figure(figsize=(18,8))

sigma  = np.linspace(.5,5,10)

for i in range(len(sigma)):
    ax = fig.add_subplot(2,5,int(i+1))
    sig = sigma[i]
    # creating a 2D Gaussian kernel with a diameter of 11 pixels
    # set with different sigmas
    gaussian_kernel = Gaussian2DKernel(sig,sig)
    # normalising the total flux of the kernel to 1
    norm = np.sum(np.array(gaussian_kernel))
    gaussian_kernel = np.array(gaussian_kernel)/norm
    # getting smoothed image using convolution
    smooth_pic1 = convolve(pic1,gaussian_kernel)
    # subtract smoothed image off of the origiinal one
    usm2 = pic1 - smooth_pic1

    # displaying the subtracted smoothed iamge
    ax.imshow(zoom(usm2, int(xcen), int(ycen), 20),cmap='cubehelix')#norm=matplotlib.colors.LogNorm()
    ax.set_title(r'$\sigma={}$'.format(sig))
    
    # END Q4 answer

    filename = './usm_gaussian/sig'+str(sig)+'.fits'
    hdu = fits.PrimaryHDU(usm2)
    hdu.writeto(filename, overwrite=True)

plt.tight_layout()
show()
```

```python

# make circular rings

def r_theta(im, xc, yc):
    '''make a radius mask'''
    # returns the radius rr and the angle phi for point (xc,yc)
    ny, nx = im.shape
    yp, xp = np.mgrid[0:ny,0:nx]
    yp = yp - yc
    xp = xp - xc
    rr = np.sqrt(np.power(yp,2.) + np.power(xp,2.))
    phi = np.arctan2(yp, xp)
    return(rr, phi)


# making the base of the figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))    

# show what r_theta() produces with a demonstration image
im1 = np.zeros((10,20))
r,t = r_theta(im1, 10,5)
ax1.imshow(r)

# creating a mask that masks everything within a radius of 6 and 9 from the central point
mask = (r > 6) * (r < 9)
ax2.imshow(mask)


# fancy plotting of the figure
ax1.set_xlabel('pixels'); ax1.set_ylabel('pixels'); ax1.set_title('r')      
ax2.set_xlabel('pixels'); ax2.set_ylabel('pixels'); ax2.set_title('(r > 6) * (r < 9)')
fig.suptitle('r, t = r_theta(im, 10, 5)', fontsize=16)        
plt.show()                         
```


<div class="alert alert-block alert-success">
<b>Example  1.4:</b> We can now make a routine that makes several concentric rings in one mask image. An example of the output is shown below. Look at the <code>for</code> loop in the routine, and look at how <code>enumerate()</code> generates index values in <code>i</code>.
</div>


**NOTE** there is a way to generate this image *without* using a <code>for</code> loop...

```python
im2 = np.zeros((40,100))

def rings(im, x, y, r_rings):
    # makes a mask for rings of different radii specified in r_rings
    # and centre (x,y)
    # invalid values are -1
    # first, second, third... ring has value (0, 1, 2, ....)
    # ring 0 is from r_rings[0] to r_rings[1]
    # ring 1 is from r_rings[1] to r_rings[2]
    im_rings = np.zeros_like(im) - 1.
    
    # make r_inner and r_outer
    r_inner = r_rings[0:-1]
    r_outer = r_rings[1:]
    
    r, t = r_theta(im_rings, x, y)
    for i, (rin, rout) in enumerate(zip(r_inner,r_outer)):
        im_rings[(r >= rin) * (r < rout)] = i

    r_middle = (r_inner + r_outer) / 2.
    return(im_rings, r_middle)

r_rings = np.array([0,10,20,30,40,50]) # 6 points, 5 bins

im_rin, r_radius = rings(im2,10,10, r_rings)
# test that rings are complete and complementary
plt.imshow(im_rin)

print ("The value r_radius contains     : ",r_radius)

nbins = r_radius.size
print ("The size of r_radius is nbins   : ",nbins)
```

The picture from example 1.4 nicely illustrates how we can make several concentric rings in one mask image.
